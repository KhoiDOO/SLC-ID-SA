```python
import tensorflow as tf
```

```python
irv2 = tf.keras.applications.InceptionResNetV2(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classifier_activation="softmax",
)
```

```python
print("Number of layer in densenet: {}".format(len(irv2.layers)))
```

```python
print("Example")
print("Layer: {}".format(irv2.layers[0]))
print("Layer input: {}".format(irv2.input))
print("Layer name: {}".format(irv2.layers[0].name))
```

```python
for i in range(len(irv2.layers)):
    print(f"Layer {i} - inverse_number: {i - len(irv2.layers)}: {irv2.layers[i].name} ")
```
