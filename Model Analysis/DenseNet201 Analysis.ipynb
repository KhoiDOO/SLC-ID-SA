```python
import tensorflow as tf
```

```python
densenet = tf.keras.applications.DenseNet201(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
)
```

```python
print("Number of layer in densenet: {}".format(len(densenet.layers)))
```

```python
print("Example")
print("Layer: {}".format(densenet.layers[0]))
print("Layer input: {}".format(densenet.input))
print("Layer name: {}".format(densenet.layers[0].name))
```

```python
for i in range(len(densenet.layers)):
    print(f"Layer {i} - inverse_number: {i - 709}: {densenet.layers[i].name} ")
```
