```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import glob
import os
```

```python
# Environment variable
main_data_dir = 'D:/Data/HAM100000 - Harvard Dataset'
img_data_dir = main_data_dir + '/img_data'
preprocessed_data = main_data_dir + '/preprocessed_data'
train_label_path = preprocessed_data + '/train_label.csv'
augmented_data_dir = main_data_dir + '/augmented_data'
test_augmented_data_dir = main_data_dir + '/test_augmented'
```

```python
train_df = pd.read_csv(train_label_path)
train_df.head()
```

```python
import tensorflow as tf

datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=180,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
```

```python
# Test augmented data
ex_img = img_data_dir + '/ISIC_0025030.jpg'
img = mpimg.imread(ex_img)
imgplot = plt.imshow(img)
plt.show()
```

```python
new_img = datagen.flow(np.array([img]), save_to_dir=test_augmented_data_dir, save_prefix='ISIC_0025030', save_format='jpg')
new_img[0][0].shape
```
